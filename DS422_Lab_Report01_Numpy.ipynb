# **10. Create the following pattern without hardcoding. Use only numpy functions and the below input array a. (np.repeat, np.tile)**

```python
a = np.array([1, 2, 3])
#> array([1, 1, 1, 2, 2, 2, 3, 3, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3])

part1 = np.repeat(a, 3)
part2 = np.tile(a, 3)
pattern = np.concatenate([part1, part2])
print(pattern)
```

# **24. Import the iris dataset keeping the text intact.**

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype='object')
names = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')
print(iris[:5])
```

# **25. Extract the text column species from the 1D iris imported in previous question.**

```python
species = iris[:, 4]
print(species[:5])
```

# **26. Convert the 1D iris to 2D array iris_2d by omitting the species text field.**

```python
iris_2d = iris[:, :4].astype(float)
print(iris_2d[:5])
```

# **27. Find the mean, median, standard deviation of iris's sepallength (1st column).**

```python
sepallength = iris_2d[:, 0]
mean_val = np.mean(sepallength)
median_val = np.median(sepallength)
std_val = np.std(sepallength)

print(f"Mean: {mean_val:.2f}, Median: {median_val:.2f}, Std Dev: {std_val:.2f}")
```

# **28. Create a normalized form of iris's sepallength whose values range exactly between 0 and 1 so that the minimum has value 0 and maximum has value 1.**

```python
S_min = sepallength.min()
S_max = sepallength.max()
sepallength_normalized = (sepallength - S_min) / (S_max - S_min)

print(sepallength_normalized[:5])
```

# **29. Find the 5th and 95th percentile of iris's sepallength.**

```python
perc_5 = np.percentile(sepallength, 5)
perc_95 = np.percentile(sepallength, 95)
print(f"5th: {perc_5:.2f}, 95th: {perc_95:.2f}")
```

# **30. Insert np.nan values at 20 random positions in iris_2d dataset.**

```python
iris_with_nan = iris_2d.copy()
num_nans = 20
rand_rows = np.random.randint(0, iris_with_nan.shape[0], num_nans)
rand_cols = np.random.randint(0, iris_with_nan.shape[1], num_nans)
iris_with_nan[rand_rows, rand_cols] = np.nan

print(f"Total NaNs inserted: {np.isnan(iris_with_nan).sum()}")
```

# **31. Find the number and position of missing values in iris_2d's sepallength.**

```python
sepallength_nan = iris_with_nan[:, 0]
nan_count = np.isnan(sepallength_nan).sum()
nan_positions = np.where(np.isnan(sepallength_nan))[0]

print(f"Number of NaNs: {nan_count}")
print(f"Positions of NaNs: {nan_positions}")
```
